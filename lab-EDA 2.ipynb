```python
#plot total number of responses

no_responses = categorical['response'].value_counts()
no_responses

# create figure and axes
# subplots()
fig1, ax1 = plt.subplots()
# set title
# set_title()
ax1.set_title("Bar plot number of responses")
# plot the bar chart
# bar()

sns.barplot(x = no_responses.index ,y = no_responses,data=categorical) #  seaborn's barplot
```

```python
#plot responses by sales channel

sns.countplot(x=categorical['sales_channel'], hue = categorical['response'])
```

```python
eda_df.response.unique()
```

```python
grouped_data = eda_df.groupby('response')['total_claim_amount'].sum().reset_index()
grouped_data
```

```python
# change colors and axes-labelling!

sns.color_palette("mako", as_cmap=True)

palette_response = {"yes": "C5", "no": "C0"}

# Start with the usual matplotlib subplots() function
fig2,axes2 = plt.subplots(1,1,figsize = (7,6))


# set the title using: set_title()
ax2.set_title("response by total claim amount")

# uses seaborn barplot(x, y, hue = , palette = ) function
sns.barplot(x = grouped_data['response'], y = grouped_data['total_claim_amount'], hue = grouped_data['response'], palette= palette_response)
```

```python
# Bar plot for 'yes' responses (Primary axis)
ax1 = sns.barplot(x='response', y='amount', data=grouped_data, ax=ax1, color='tab:blue', label='Total Amount')
ax1.set_ylabel('Total Amount', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
```

```python
group_income = eda_df.groupby('response')['income'].sum().reset_index()
group_income
```

```python
# change colors and axes-labelling!

sns.color_palette("mako", as_cmap=True)

palette_response = {"yes": "C5", "no": "C0"}

# Start with the usual matplotlib subplots() function
fig2,axes2 = plt.subplots(1,1,figsize = (7,6))


# set the title using: set_title()
ax2.set_title("response by income")

# uses seaborn barplot(x, y, hue = , palette = ) function
sns.barplot(x = group_income['response'], y = group_income['income'], hue = group_income['response'], palette= palette_response)
```

### optional - Findings in dataset

1. Multicollinearity: correlation between explanatory variables <90% - probably no multicollinearity --> variables seem ti have significant effect on dependent variable

2. Overfitting: correlation below <90%: no overfitting of model, model should be good for other data sets as well

3. there is a positive relation between the amount a person pays for their insurance and the amount of claims they make: for every claim that is made, the premium rises by 0.63 percentage points.



```python
sns.set_style("whitegrid")
g = sns.lmplot(x="total_claim_amount", y="monthly_premium_auto",data=eda_df)
g = (g.set_axis_labels(x="total claim amount",y="monthly premium auto").set(xlim=(0,3500),ylim=(40,400)))
plt.title("positive relation between total claim amount and monthly premium")
plt.show(g)
```

```python
sns.set_style("whitegrid")
g = sns.lmplot(x="monthly_premium_auto", y="total_claim_amount",data=eda_df)
g = (g.set_axis_labels(x="monthly premium auto",y="total_claim_amount").set(xlim=(40,400),ylim=(0,3500)))
plt.title("positive relation between total claim amount and monthly premium")
plt.show(g)
```
