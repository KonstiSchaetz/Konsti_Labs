```python
#plot total number of responses

no_responses = categorical['response'].value_counts()
no_responses

# create figure and axes
# subplots()
fig1, ax1 = plt.subplots()
# set title
# set_title()
ax1.set_title("Bar plot number of responses")
# plot the bar chart
# bar()

sns.barplot(x = no_responses.index ,y = no_responses,data=categorical) #  seaborn's barplot
```

```python
#plot responses by sales channel

sns.countplot(x=categorical['sales_channel'], hue = categorical['response'])
```

```python
# plot resonses by total claim amount

sns.catplot(data=eda_df, x="response", y="total_claim_amount")
```

```python
eda_df.response.unique()
```

```python
grouped_data = eda_df.groupby('response')['total_claim_amount'].sum().reset_index()
grouped_data
```

```python
# change colors and axes-labelling!

sns.color_palette("mako", as_cmap=True)

palette_response = {"yes": "C5", "no": "C0"}

# Start with the usual matplotlib subplots() function
fig2,axes2 = plt.subplots(1,1,figsize = (7,6))


# set the title using: set_title()
ax2.set_title("response by total claim amount")

# uses seaborn barplot(x, y, hue = , palette = ) function
sns.barplot(x = grouped_data['response'], y = grouped_data['total_claim_amount'], hue = grouped_data['response'], palette= palette_response)
```

```python
group_income = eda_df.groupby('response')['income'].sum().reset_index()
group_income
```

```python
# change colors and axes-labelling!

sns.color_palette("mako", as_cmap=True)

palette_response = {"yes": "C5", "no": "C0"}

# Start with the usual matplotlib subplots() function
fig2,axes2 = plt.subplots(1,1,figsize = (7,6))


# set the title using: set_title()
ax2.set_title("response by income")

# uses seaborn barplot(x, y, hue = , palette = ) function
sns.barplot(x = group_income['response'], y = group_income['income'], hue = group_income['response'], palette= palette_response)
```
